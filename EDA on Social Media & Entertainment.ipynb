## Distribution of Daily Minutes_Spent

```python
plt.figure(figsize=(6,4))

sns.histplot(df['Daily_Minutes_Spent'], bins=30, kde=True, color='blue')
plt.title('Distribution of Daily Minutes_Spent')
plt.xlabel('Minutes Spent')
plt.ylabel('No.of Users')
plt.show()
```

## Distribution of Posts Per Day

```python
plt.figure(figsize=(6,4))

sns.histplot(df['Posts_Per_Day'], bins=20, kde=True, color='green')
plt.title('Distribution of Posts Per Day')
plt.xlabel('Posts per Day')
plt.ylabel('No. of Users')
plt.show()
```

## Distribution of Likes per Day

```python
plt.figure(figsize=(6,4))

sns.histplot(df['Likes_Per_Day'], bins=30, kde=True, color='red')
plt.title('Distribution of Likes per Day')
plt.xlabel('Likes per Day')
plt.ylabel('No. of Users')
plt.show()
```

## Distribution of Follows per Day

```python
plt.figure(figsize=(6,4))

sns.histplot(df['Follows_Per_Day'], bins=30, kde=True, color='purple')
plt.title('Distribution of Follows per Day')
plt.xlabel('Follows per Day')
plt.ylabel('No.of Users')
plt.show()
```

## **Insights from Distribution Plots:**

- Daily Minutes Spent: Most users spend between 100-400 minutes per day, with a peak around 250 minutes.

- Posts Per Day: The distribution is right-skewed, meaning a majority post fewer times, with fewer users posting 15-20 times daily.

- Likes Per Day: A normal-like distribution centered around 95 likes per day, with some users liking up to 200 posts.

- Follows Per Day: Most users follow around 25 accounts daily, but some users follow none.

## Daily Minutes Spent by App

```python
plt.figure(figsize=(6,4))

sns.boxplot(x='App', y='Daily_Minutes_Spent', data=df, palette="Set2")
plt.title('Daily Minutes Spent by App')
plt.xticks(rotation=45)
plt.show()
```

## Posts Per Day by App

```python
plt.figure(figsize=(6,4))

sns.boxplot(x='App', y='Posts_Per_Day', data=df , palette="Set3")
plt.title('Posts Per Day by App')
plt.xticks(rotation=45)
plt.show()
```

## Likes Per Day by App

```python
plt.figure(figsize=(6,4))

sns.boxplot(x='App', y='Likes_Per_Day', data=df, palette='coolwarm')
plt.title('Likes Per Day by App')
plt.xticks(rotation=45)
plt.show()
```

## Follows Per Day by App

```python
plt.figure(figsize=(6,4))

sns.boxplot(x='App', y='Follows_Per_Day', data=df, palette='viridis')
plt.title('Follows Per Day by App')
plt.xticks(rotation=45)
plt.show()
```

# **Insights from Boxplots:**

**1.Daily Minutes Spent:**

TikTok & Instagram have the highest median usage time, with many users spending 300+ minutes daily.
LinkedIn & Twitter show lower engagement, with most users spending <250 minutes.

**2.Posts Per Day:**

TikTok and Snapchat have higher median posts per day, while LinkedIn & Twitter show lower posting activity.
Outliers indicate that some users post significantly more frequently on these platforms.

**3.Likes Per Day:**

Instagram, TikTok, and Facebook see the highest daily likes.
LinkedIn has the lowest median likes, suggesting lower engagement in terms of liking content.

**4.Follows Per Day:**

Instagram and Twitter users follow more accounts daily, aligning with their nature of content discovery.
Pinterest and LinkedIn have lower follow activity.

```python
df_temp = df.drop(columns=['User_ID', 'App'])
correlation_matrix = df_temp.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Social Media Usage Features")
plt.show()
```

   - Has a very weak negative correlation with **Posts_Per_Day** (-0.02).

2. **Posts_Per_Day**:
   - Has a very weak positive correlation with **Likes_Per_Day** (0.02) and **Follows_Per_Day** (0.02).
   - Has a very weak negative correlation with **Daily_Minutes_Spent** (-0.02).

3. **Likes_Per_Day**:
   - Has a very weak positive correlation with **Daily_Minutes_Spent** (0.01) and **Posts_Per_Day** (0.02).
   - Has a very weak negative correlation with **Follows_Per_Day** (-0.02).

4. **Follows_Per_Day**:
   - Has a very weak positive correlation with **Daily_Minutes_Spent** (0.04) and **Posts_Per_Day** (0.02).
   - Has a very weak negative correlation with **Likes_Per_Day** (-0.02).

Overall, these correlations are very weak, indicating that there is not much of a relationship between these different aspects of social media usage. This insight can help in understanding user behavior and informing social media strategies.
